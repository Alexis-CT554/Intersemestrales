### Sistema 3
#### Este ejercicio si tiene solucion ya los planos de las ecuaciones se intersectan 

```python
y = np.linspace(-12, 8)
x2, y2 = np.meshgrid(x, y)
z1 = (3-2*x2+y2)/2
z2 = (-2*x2-3*y2)/-1
z3 = (-12-x2-2*y2)/2
fig = plt.figure()
ax = fig.add_subplot(projection="3d") 
ax.plot_surface(x2, y2, z1)
ax.plot_surface(x2, y2, z2)
ax.plot_surface(x2, y2, z3)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("SISTEMA 3")
ax.view_init(elev=0, azim=18)
plt.show()
```
